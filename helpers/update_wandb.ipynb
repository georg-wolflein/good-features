```python
import wandb

api = wandb.Api()
runs = [run for run in api.runs("histaug", order="+created_at", per_page=1000)]
```

```python
from tqdm import tqdm

# Iterate over all runs
for run in tqdm(runs):
    # Update the config value
    if "magnification" in run.config:
        del run.config["magnification"]
    run.config["settings"]["magnification"] = run.config["settings"].get("magnification", "low")
    # Save the changes
    run.update()
```
